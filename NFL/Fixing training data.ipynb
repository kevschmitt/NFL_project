```python
import pandas as pd
df = pd.read_csv('data/spreadspoke_scores.csv')
```

```python
columns = ['schedule_date', 'team_home', 'team_away', 'net_score', 'total_points', 'spread_favorite', 'over_under_line']
```

```python
df = df[df.schedule_date == '1/3/2021']
```

```python
df.head()
```

```python
df['net_score'] = df['score_home'] - df['score_away']
df['total_points'] = df['score_home'] + df['score_away']
df = df[columns]
```

```python
df.to_csv('data/delete.csv', index=False)
```
